# Radon Transform

```python
import numpy as np
import matplotlib.pyplot as plt

from skimage.io import imread
from skimage.data import shepp_logan_phantom
from skimage.transform import radon, rescale
import pandas as pd
from skimage.transform import iradon
```

```python
img = imread('baw.jpg')
gray_img = img[:,:,0]
plt.imshow(gray_img, cmap = "Greys_r")
```

## Sinogram with angle 0

```python
sinogram = radon(gray_img, theta=[0.], circle=True)
plt.plot(sinogram)
```

## Sinogram with angle 90

```python
sinogram = radon(gray_img, theta=[90.], circle=True)
plt.plot(sinogram)
```

## Sinograms in angles between 0 to 180

```python
theta = np.linspace(0., 180., 500, endpoint=False)
sinogram = radon(gray_img, theta=theta, circle=True)
plt.imshow(sinogram, cmap = "Greys_r")
```

```python
theta = np.linspace(0., 180., max(amoeba.shape), endpoint = False)
si = radon(amoeba, theta = theta, circle = False)
plt.imshow(si, cmap = "Greys_r")
```

```python
inverted_image = iradon(si, theta = theta, circle = False)
plt.imshow(inverted_image, cmap = "Greys_r")
```
