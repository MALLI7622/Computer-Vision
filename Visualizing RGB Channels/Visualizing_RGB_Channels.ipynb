```
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import image as mpimg
import cv2

%matplotlib inline
```

```
image = mpimg.imread('wa_state_highway.png')

plt.imshow(image)
```

```

r = image[:,:,0]

g = image[:,:,1]

b = image[:,:,2]

f, (ax1, ax2, ax3) = plt.subplots(1,3, figsize = (20,10))

ax1.set_title('R Channel')
ax1.imshow(r, cmap='Greys_r')
ax2.set_title('G Channel  ')
ax2.imshow(g, cmap='Greys_r')
ax3.set_title('B Channel')
ax3.imshow(b, cmap='Greys_r')
```
