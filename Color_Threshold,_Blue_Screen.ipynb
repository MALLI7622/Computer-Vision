```
from matplotlib import pyplot as plt
from matplotlib import image as mpimg
import numpy as np
import cv2
%matplotlib inline
```

```
image = cv2.imread('waymo.jpg')
print('The image is:',type(image),
      'Image shape',image.shape)


plt.imshow(image, cmap='viridis')
```

```
image_copy = np.copy(image)

image_copy = cv2.cvtColor(image_copy,cv2.COLOR_BGR2RGB)

plt.imshow(image_copy)
```

```
lower_blue = np.array([0,0,230])
upper_blue = np.array([250,250,255])
```

```
upper_blue
```

```
mask = cv2.inRange(image_copy,lower_blue,upper_blue)

plt.imshow(mask,cmap = 'gray')
```

```
masked_image = np.copy (image_copy)

masked_image[mask!=0] = [0,0,0]

plt.imshow(masked_image)
```

```
mask != 0
```

```
background_image = cv2.imread('space.jpg')

plt.imshow(background_image)
```

```
background_image = cv2.cvtColor(background_image, cv2.COLOR_BGR2RGB)
plt.imshow(background_image)
```

```
background_image.shape
```

```
crop_background = background_image[0:84,0:150]

crop_background[mask == 0] = [0,0,0]

plt.imshow(crop_background)
```

```
complete_image = image_copy + background_image

plt.imshow(complete_image)
```
