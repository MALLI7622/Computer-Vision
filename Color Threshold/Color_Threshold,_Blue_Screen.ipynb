```
from matplotlib import pyplot as plt
import cv2
import numpy as np
```

```
image = cv2.imread('pizza_bluescreen.jpg')

print("Image type:",type(image),
      "Image Size",image.shape)
```

```
%matplotlib inline
plt.imshow(image)
```

```
image_copy = np.copy(image)
```

```

image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)

plt.imshow(image_copy)
```

```
 lower_blue = np.array([0,0, 220])

 upper_blue = np.array([50,70,255])
```

```
mask = cv2.inRange(image_copy, lower_blue,upper_blue)

plt.imshow(mask, cmap = "Greys_r")
```

```
masked_copy = np.copy(image_copy)

masked_copy[mask != 0 ] = [0,0,0]

plt.imshow(masked_copy)
```

```
background_image = cv2.imread("space_background.jpg")
background_image = cv2.cvtColor(background_image, cv2.COLOR_BGR2RGB)

crop_background = background_image[0:514, 0:816]

crop_background[mask == 0 ] = [0,0,0]

plt.imshow(crop_background)
```

```
complete_image = crop_background + masked_copy

plt.imshow(complete_image)
```
