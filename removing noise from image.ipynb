```python
n = np.shape(pandu1)
```

```python
print(*n)
```

```python
b = np.multiply(np.random.rand(*n),noise_sigma)
```

```python
b[1,1,1]
```

```python
np.shape(pandu)
```

```python
pandu_noise = pandu + b
```

```python
blur = cv2.blur(pandu1,(25,25))
```

```python
cv2.imshow('frame',(pandu_noise).astype(np.uint8))
cv2.waitKey(0)
cv2.destroyAllWindows()
```
